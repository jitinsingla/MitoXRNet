```python
import os,gc,glob
import sys
import PathConfig
import numpy as np
from patchify import patchify
from read_write_mrc import read_mrc, write_mrc
from utils import create_and_save_slices
```

```python
subFolders = sorted(os.listdir('../Data/Preprocessed_Data/'))

for i in subFolders:    
    path_to_subfolder = os.path.join('../Data/Preprocessed_Data/' , i)
    output_folder = os.path.join('../Data/Training_Slices/' , i+'_Slices')
    os.makedirs(output_folder,exist_ok=True)

    for filename in os.listdir(path_to_subfolder):
        if filename.endswith('.mrc'):  
            image_path = os.path.join(path_to_subfolder, filename)

            image = read_mrc(image_path).get()

            img_name, _ = os.path.splitext(filename)

            create_and_save_slices(image, img_name,output_folder)
            del img_name, image, image_path
            gc.collect()
            
```
